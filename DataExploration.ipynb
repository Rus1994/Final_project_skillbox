```python
import pandas as pd
import numpy as np
```

```python
target_actions = ['sub_car_claim_click', 'sub_car_claim_submit_click', 
'sub_open_dialog_click', 'sub_custom_question_submit_click', 
'sub_call_number_click', 'sub_callback_submit_click', 'sub_submit_success', 
'sub_car_request_submit_click']
```

```python
df = pd.read_csv(r'data/ga_hits.csv', delimiter=',', usecols=['session_id', 'event_action'])
```

```python
df_session = pd.read_csv(r'data/ga_sessions.csv', delimiter=',')
```

```python
df.session_id.nunique()
```

```python
df['is_target_action'] = df.event_action.map(lambda action: action in target_actions)
```

```python
df.drop('event_action', axis=1, inplace=True)
df
```

```python
df_without_dupl = df.drop_duplicates()
df_without_dupl.shape
```

```python
df_without_dupl.is_target_action.value_counts()
```

```python
df_true = df_without_dupl[df_without_dupl.is_target_action]
df_true
```

```python
df_false = df_without_dupl[df_without_dupl.is_target_action == False]
df_false
```

```python
df_true_list = df_true.session_id.to_list()
```

```python
df_false_list = df_false.session_id.to_list() # удалить значения df_true_list
```

```python
all_sessions_id = set(df_false_list)
len(all_sessions_id)
```

```python
all_sessions_id = all_sessions_id.union(set(df_true_list))
```

```python
df_session['is_target_action'] = df_session.session_id.map(lambda s_id: -1 if s_id not in all_sessions_id else (1 if s_id in df_true_list else 0))
```

```python
df_session.is_target_action.value_counts()
```

```python
df_session.to_csv(r'data/sessions.csv')
```
